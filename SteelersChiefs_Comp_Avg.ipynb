```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
steelers_csvs = os.listdir('./Steelers_Chiefs/CSVs/Steelers_CSVs')
steelers_dfs = [pd.read_csv('./Steelers_Chiefs/CSVs/Steelers_CSVs/'+csv) for csv in steelers_csvs]
chiefs_csvs = os.listdir('./Steelers_Chiefs/CSVs/Chiefs_CSVs')
chiefs_dfs = [pd.read_csv('./Steelers_Chiefs/CSVs/Chiefs_CSVs/'+csv) for csv in chiefs_csvs]
```

```python
steelers_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in steelers_dfs]
steelers_dfs_avg_comps_reordered = [steelers_dfs_avg_comps[11],steelers_dfs_avg_comps[12],steelers_dfs_avg_comps[13],steelers_dfs_avg_comps[2],steelers_dfs_avg_comps[3],steelers_dfs_avg_comps[4], steelers_dfs_avg_comps[5], steelers_dfs_avg_comps[6], steelers_dfs_avg_comps[7], steelers_dfs_avg_comps[8], steelers_dfs_avg_comps[9], steelers_dfs_avg_comps[10],steelers_dfs_avg_comps[0], steelers_dfs_avg_comps[1]]
steelers_dfs_win_perc = [df.loc[df['Positive_Sentiment']==True].shape[0]/df.shape[0] for df in steelers_dfs]
steelers_dfs_win_percs_reordered = [steelers_dfs_win_perc[11],steelers_dfs_win_perc[12],steelers_dfs_win_perc[13],steelers_dfs_win_perc[2],steelers_dfs_win_perc[3],steelers_dfs_win_perc[4], steelers_dfs_win_perc[5], steelers_dfs_win_perc[6], steelers_dfs_win_perc[7], steelers_dfs_win_perc[8], steelers_dfs_win_perc[9], steelers_dfs_win_perc[10],steelers_dfs_win_perc[0], steelers_dfs_win_perc[1]]
steelers_wl = ['win', 'win', 'win', 'win', 'loss', 'loss', 'loss', 'loss', 'win', 'loss', 'loss', 'win', 'loss', 'win']

chiefs_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in chiefs_dfs]
chiefs_dfs_avg_comps_reordered = [chiefs_dfs_avg_comps[11],chiefs_dfs_avg_comps[12],chiefs_dfs_avg_comps[13],chiefs_dfs_avg_comps[2],chiefs_dfs_avg_comps[3],chiefs_dfs_avg_comps[4], chiefs_dfs_avg_comps[5], chiefs_dfs_avg_comps[6], chiefs_dfs_avg_comps[7], chiefs_dfs_avg_comps[8], chiefs_dfs_avg_comps[9], chiefs_dfs_avg_comps[10],chiefs_dfs_avg_comps[0], chiefs_dfs_avg_comps[1]]
chiefs_dfs_win_perc = [df.loc[df['Positive_Sentiment']==True].shape[0]/df.shape[0] for df in chiefs_dfs]
chiefs_dfs_win_percs_reordered = [chiefs_dfs_win_perc[11],chiefs_dfs_win_perc[12],chiefs_dfs_win_perc[13],chiefs_dfs_win_perc[2],chiefs_dfs_win_perc[3],chiefs_dfs_win_perc[4], chiefs_dfs_win_perc[5], chiefs_dfs_win_perc[6], chiefs_dfs_win_perc[7], chiefs_dfs_win_perc[8], chiefs_dfs_win_perc[9], chiefs_dfs_win_perc[10],chiefs_dfs_win_perc[0], chiefs_dfs_win_perc[1]]
chiefs_wl = ['loss', 'win', 'win', 'win', 'win', 'loss', 'win', 'win', 'loss', 'win', 'win', 'win', 'win', 'loss']
```

```python
fig, (steelers_plot_1, steelers_plot_2) = plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(ax=steelers_plot_1 , x=range(len(steelers_dfs_avg_comps))+np.ones(len(steelers_dfs_avg_comps)), y=steelers_dfs_avg_comps_reordered, hue=steelers_wl)
steelers_plot_1.set_title('Steelers Avg. Sentiment Per Game')
steelers_plot_1.set_xlabel('Game Number')
steelers_plot_1.set_xticks(range(len(steelers_dfs_avg_comps))+np.ones(len(steelers_dfs_avg_comps)))
steelers_plot_1.set_ylabel('Comp. Sentiment')

sns.scatterplot(ax=steelers_plot_2 , x=range(len(steelers_dfs_win_perc))+np.ones(len(steelers_dfs_win_perc)), y=steelers_dfs_win_percs_reordered, hue=steelers_wl)
steelers_plot_2.set_title('Steelers Positive Perc. Per Game')
steelers_plot_2.set_xlabel('Game Number')
steelers_plot_2.set_xticks(range(len(steelers_dfs_win_perc))+np.ones(len(steelers_dfs_win_perc)))
steelers_plot_2.set_ylabel('Positive Perc.')
```

```python
fig, (chiefs_plot_1, chiefs_plot_2) = plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(ax=chiefs_plot_1 , x=range(len(chiefs_dfs_avg_comps))+np.ones(len(chiefs_dfs_avg_comps)), y=chiefs_dfs_avg_comps_reordered, hue=chiefs_wl)
chiefs_plot_1.set_title('Chiefs Avg. Sentiment Per Game')
chiefs_plot_1.set_xlabel('Game Number')
chiefs_plot_1.set_xticks(range(len(chiefs_dfs_avg_comps))+np.ones(len(chiefs_dfs_avg_comps)))
chiefs_plot_1.set_ylabel('Comp. Sentiment')

sns.scatterplot(ax=chiefs_plot_2 , x=range(len(chiefs_dfs_win_perc))+np.ones(len(chiefs_dfs_win_perc)), y=chiefs_dfs_win_percs_reordered, hue=chiefs_wl)
chiefs_plot_2.set_title('Chiefs Positive Perc. Per Game')
chiefs_plot_2.set_xlabel('Game Number')
chiefs_plot_2.set_xticks(range(len(chiefs_dfs_win_perc))+np.ones(len(chiefs_dfs_win_perc)))
chiefs_plot_2.set_ylabel('Positive Perc.')
```
