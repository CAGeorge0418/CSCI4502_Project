```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
niners_csvs = os.listdir('Niners_CSVs')
niners_dfs = [pd.read_csv('Niners_CSVs/'+csv) for csv in niners_csvs]
bears_csvs = os.listdir('Bears_CSVs')
bears_dfs = [pd.read_csv('Bears_CSVs/'+csv) for csv in bears_csvs]
```

```python
niners_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in niners_dfs]
niners_dfs_avg_comps_reordered = [niners_dfs_avg_comps[9],niners_dfs_avg_comps[5],niners_dfs_avg_comps[0],niners_dfs_avg_comps[3],niners_dfs_avg_comps[10],niners_dfs_avg_comps[1], niners_dfs_avg_comps[7], niners_dfs_avg_comps[8], niners_dfs_avg_comps[6], niners_dfs_avg_comps[4], niners_dfs_avg_comps[2]]
niners_dfs_win_perc = [df.loc[df['Positive_Sentiment']==True].shape[0]/df.shape[0] for df in niners_dfs]
niners_dfs_win_percs_reordered = [niners_dfs_win_perc[9],niners_dfs_win_perc[5],niners_dfs_win_perc[0],niners_dfs_win_perc[3],niners_dfs_win_perc[10],niners_dfs_win_perc[1], niners_dfs_win_perc[7], niners_dfs_win_perc[8], niners_dfs_win_perc[6], niners_dfs_win_perc[4], niners_dfs_win_perc[2]]
niners_wl = ['loss', 'win', 'loss', 'win', 'win', 'loss', 'loss', 'win', 'win', 'win', 'win']

bears_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in bears_dfs]
bears_dfs_avg_comps_reordered = [bears_dfs_avg_comps[7],bears_dfs_avg_comps[10],bears_dfs_avg_comps[11], bears_dfs_avg_comps[12],bears_dfs_avg_comps[6],bears_dfs_avg_comps[9], bears_dfs_avg_comps[5], bears_dfs_avg_comps[3], bears_dfs_avg_comps[13], bears_dfs_avg_comps[4], bears_dfs_avg_comps[8], bears_dfs_avg_comps[2],bears_dfs_avg_comps[0], bears_dfs_avg_comps[1]]
bears_dfs_win_perc = [df.loc[df['Positive_Sentiment']==True].shape[0]/df.shape[0] for df in bears_dfs]
bears_dfs_win_percs_reordered = [bears_dfs_win_perc[7],bears_dfs_win_perc[10],bears_dfs_win_perc[11], bears_dfs_win_perc[12],bears_dfs_win_perc[6],bears_dfs_win_perc[9], bears_dfs_win_perc[5], bears_dfs_win_perc[3], bears_dfs_win_perc[13], bears_dfs_win_perc[4], bears_dfs_win_perc[8], bears_dfs_win_perc[2],bears_dfs_win_perc[0], bears_dfs_win_perc[1]]
bears_wl = ['win', 'win', 'win', 'win', 'loss', 'win', 'loss', 'loss', 'loss', 'win', 'loss', 'loss', 'loss', 'loss']
```

```python
fig, (niners_plot_1, niners_plot_2) = plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(ax=niners_plot_1 , x=range(len(niners_dfs_avg_comps))+np.ones(len(niners_dfs_avg_comps)), y=niners_dfs_avg_comps_reordered, hue=niners_wl)
niners_plot_1.set_title('Niners Average Sentiment Per Game')
niners_plot_1.set_xlabel('Game Number')
niners_plot_1.set_xticks(range(len(niners_dfs_avg_comps))+np.ones(len(niners_dfs_avg_comps)))
niners_plot_1.set_ylabel('Compound Sentiment')

sns.scatterplot(ax=niners_plot_2 , x=range(len(niners_dfs_win_perc))+np.ones(len(niners_dfs_win_perc)), y=niners_dfs_win_percs_reordered, hue=niners_wl)
niners_plot_2.set_title('Niners Positive Percentage Per Game')
niners_plot_2.set_xlabel('Game Number')
niners_plot_2.set_xticks(range(len(niners_dfs_win_perc))+np.ones(len(niners_dfs_win_perc)))
niners_plot_2.set_ylabel('Positive Percentage')
```

```python
fig, (bears_plot_1, bears_plot_2) = plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(ax=bears_plot_1 , x=range(len(bears_dfs_avg_comps))+np.ones(len(bears_dfs_avg_comps)), y=bears_dfs_avg_comps_reordered, hue=bears_wl, palette=['orange','steelblue'])
bears_plot_1.set_title('Bears Average Sentiment Per Game')
bears_plot_1.set_xlabel('Game Number')
bears_plot_1.set_xticks(range(len(bears_dfs_avg_comps))+np.ones(len(bears_dfs_avg_comps)))
bears_plot_1.set_ylabel('Compound Sentiment')

sns.scatterplot(ax=bears_plot_2 , x=range(len(bears_dfs_win_perc))+np.ones(len(bears_dfs_win_perc)), y=bears_dfs_win_percs_reordered, hue=bears_wl, palette=['orange','steelblue'])
bears_plot_2.set_title('Bears Positive Percentage Per Game')
bears_plot_2.set_xlabel('Game Number')
bears_plot_2.set_xticks(range(len(bears_dfs_win_perc))+np.ones(len(bears_dfs_win_perc)))
bears_plot_2.set_ylabel('Positive Percentage')
```
