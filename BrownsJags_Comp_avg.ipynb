```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
browns_csvs = os.listdir('Browns_CSVs')
browns_dfs = [pd.read_csv('Browns_CSVs/'+csv) for csv in browns_csvs]
jags_csvs = os.listdir('Jaguars_CSVs')
jags_dfs = [pd.read_csv('Jaguars_CSVs/'+csv) for csv in jags_csvs]
```

```python
browns_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in browns_dfs]
browns_dfs_avg_comps_reordered = [browns_dfs_avg_comps[11],browns_dfs_avg_comps[12],browns_dfs_avg_comps[13],browns_dfs_avg_comps[2],browns_dfs_avg_comps[3],browns_dfs_avg_comps[4], browns_dfs_avg_comps[5], browns_dfs_avg_comps[6], browns_dfs_avg_comps[7], browns_dfs_avg_comps[8], browns_dfs_avg_comps[9], browns_dfs_avg_comps[10],browns_dfs_avg_comps[0], browns_dfs_avg_comps[1]]
browns_wl = ['win', 'loss', 'loss', 'win', 'loss', 'win', 'loss', 'loss', 'loss', 'loss', 'win', 'loss', 'loss', 'win']

jags_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in jags_dfs]
jags_dfs_avg_comps_reordered = [jags_dfs_avg_comps[11],jags_dfs_avg_comps[12],jags_dfs_avg_comps[13],jags_dfs_avg_comps[2],jags_dfs_avg_comps[3],jags_dfs_avg_comps[4], jags_dfs_avg_comps[5], jags_dfs_avg_comps[6], jags_dfs_avg_comps[7], jags_dfs_avg_comps[8], jags_dfs_avg_comps[9], jags_dfs_avg_comps[10],jags_dfs_avg_comps[0], jags_dfs_avg_comps[1]]
jags_wl = ['win', 'loss', 'loss', 'win', 'loss', 'win', 'loss', 'loss', 'loss', 'loss', 'win', 'loss', 'loss', 'win']
```

```python
browns_plot = sns.scatterplot(x=range(len(browns_dfs_avg_comps))+np.ones(len(browns_dfs_avg_comps)), y=browns_dfs_avg_comps_reordered, hue=browns_wl)
browns_plot.set_title('Browns Avg. Sentiment Per Game')
browns_plot.set_xlabel('Game Number')
browns_plot.set_xticks(range(len(browns_dfs_avg_comps))+np.ones(len(browns_dfs_avg_comps)))
browns_plot.set_ylabel('Comp. Sentiment')
```

```python
jags_plot = sns.scatterplot(x=range(len(jags_dfs_avg_comps))+np.ones(len(jags_dfs_avg_comps)), y=jags_dfs_avg_comps_reordered, hue=jags_wl)
jags_plot.set_title('Jags Avg. Sentiment Per Game')
jags_plot.set_xlabel('Game Number')
jags_plot.set_xticks(range(len(jags_dfs_avg_comps))+np.ones(len(jags_dfs_avg_comps)))
jags_plot.set_ylabel('Comp. Sentiment')
```
