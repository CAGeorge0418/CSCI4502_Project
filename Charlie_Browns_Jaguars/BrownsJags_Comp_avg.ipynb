```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
browns_csvs = os.listdir('Browns/Browns_CSVs')
browns_dfs = [pd.read_csv('Browns/Browns_CSVs/'+csv) for csv in browns_csvs]
jags_csvs = os.listdir('Jaguars/Jaguars_CSVs')
jags_dfs = [pd.read_csv('Jaguars/Jaguars_CSVs/'+csv) for csv in jags_csvs]
```

```python
browns_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in browns_dfs]
browns_dfs_avg_comps_reordered = [browns_dfs_avg_comps[11],browns_dfs_avg_comps[12],browns_dfs_avg_comps[13],browns_dfs_avg_comps[2],browns_dfs_avg_comps[3],browns_dfs_avg_comps[4], browns_dfs_avg_comps[5], browns_dfs_avg_comps[6], browns_dfs_avg_comps[7], browns_dfs_avg_comps[8], browns_dfs_avg_comps[9], browns_dfs_avg_comps[10],browns_dfs_avg_comps[0], browns_dfs_avg_comps[1]]
browns_dfs_win_perc = [df.loc[df['Positive_Sentiment']==True].shape[0]/df.shape[0] for df in browns_dfs]
browns_dfs_win_percs_reordered = [browns_dfs_win_perc[11],browns_dfs_win_perc[12],browns_dfs_win_perc[13],browns_dfs_win_perc[2],browns_dfs_win_perc[3],browns_dfs_win_perc[4], browns_dfs_win_perc[5], browns_dfs_win_perc[6], browns_dfs_win_perc[7], browns_dfs_win_perc[8], browns_dfs_win_perc[9], browns_dfs_win_perc[10],browns_dfs_win_perc[0], browns_dfs_win_perc[1]]
browns_wl = ['win', 'loss', 'loss', 'win', 'loss', 'win', 'loss', 'loss', 'loss', 'loss', 'win', 'loss', 'loss', 'win']

jags_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in jags_dfs]
jags_dfs_avg_comps_reordered = [jags_dfs_avg_comps[11],jags_dfs_avg_comps[12],jags_dfs_avg_comps[13],jags_dfs_avg_comps[2],jags_dfs_avg_comps[3],jags_dfs_avg_comps[4], jags_dfs_avg_comps[5], jags_dfs_avg_comps[6], jags_dfs_avg_comps[7], jags_dfs_avg_comps[8], jags_dfs_avg_comps[9], jags_dfs_avg_comps[10],jags_dfs_avg_comps[0], jags_dfs_avg_comps[1]]
jags_dfs_win_perc = [df.loc[df['Positive_Sentiment']==True].shape[0]/df.shape[0] for df in jags_dfs]
jags_dfs_win_percs_reordered = [jags_dfs_win_perc[11],jags_dfs_win_perc[12],jags_dfs_win_perc[13],jags_dfs_win_perc[2],jags_dfs_win_perc[3],jags_dfs_win_perc[4], jags_dfs_win_perc[5], jags_dfs_win_perc[6], jags_dfs_win_perc[7], jags_dfs_win_perc[8], jags_dfs_win_perc[9], jags_dfs_win_perc[10],jags_dfs_win_perc[0], jags_dfs_win_perc[1]]
jags_wl = ['loss', 'loss', 'loss', 'loss', 'win', 'win', 'loss', 'loss', 'loss', 'loss', 'loss', 'win', 'loss', 'win']
```

```python
fig, (browns_plot_1, browns_plot_2) = plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(ax=browns_plot_1 , x=range(len(browns_dfs_avg_comps))+np.ones(len(browns_dfs_avg_comps)), y=browns_dfs_avg_comps_reordered, hue=browns_wl, palette=['orange','steelblue'])
browns_plot_1.set_title('Browns Avg. Sentiment Per Game')
browns_plot_1.set_xlabel('Game Number')
browns_plot_1.set_xticks(range(len(browns_dfs_avg_comps))+np.ones(len(browns_dfs_avg_comps)))
browns_plot_1.set_ylabel('Comp. Sentiment')

sns.scatterplot(ax=browns_plot_2 , x=range(len(browns_dfs_win_perc))+np.ones(len(browns_dfs_win_perc)), y=browns_dfs_win_percs_reordered, hue=browns_wl, palette=['orange','steelblue'])
browns_plot_2.set_title('Browns Positive Perc. Per Game')
browns_plot_2.set_xlabel('Game Number')
browns_plot_2.set_xticks(range(len(browns_dfs_win_perc))+np.ones(len(browns_dfs_win_perc)))
browns_plot_2.set_ylabel('Positive Perc.')
```

```python
fig, (jags_plot_1, jags_plot_2) = plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(ax=jags_plot_1 , x=range(len(jags_dfs_avg_comps))+np.ones(len(jags_dfs_avg_comps)), y=jags_dfs_avg_comps_reordered, hue=jags_wl, palette=['steelblue','orange'])
jags_plot_1.set_title('Jaguars Avg. Sentiment Per Game')
jags_plot_1.set_xlabel('Game Number')
jags_plot_1.set_xticks(range(len(jags_dfs_avg_comps))+np.ones(len(jags_dfs_avg_comps)))
jags_plot_1.set_ylabel('Comp. Sentiment')

sns.scatterplot(ax=jags_plot_2 , x=range(len(jags_dfs_win_perc))+np.ones(len(jags_dfs_win_perc)), y=jags_dfs_win_percs_reordered, hue=jags_wl, palette=['steelblue','orange'])
jags_plot_2.set_title('Jaguars Positive Perc. Per Game')
jags_plot_2.set_xlabel('Game Number')
jags_plot_2.set_xticks(range(len(jags_dfs_win_perc))+np.ones(len(jags_dfs_win_perc)))
jags_plot_2.set_ylabel('Positive Perc.')
```
