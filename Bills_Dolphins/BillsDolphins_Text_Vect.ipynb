```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bills_csvs = os.listdir('Bills_CSVs')
bills_dfs = [pd.read_csv('Bills_CSVs/'+csv) for csv in bills_csvs]
dolphs_csvs = os.listdir('Dolphins_CSVs')
dolphs_dfs = [pd.read_csv('Dolphins_CSVs/'+csv) for csv in dolphs_csvs]
for df in bills_dfs:
    df['Date'] = pd.to_datetime(df['Date'])
for df in dolphs_dfs:
    df['Date'] = pd.to_datetime(df['Date'])

mega_Bills_df = pd.concat(bills_dfs)
mega_Dolphs_df = pd.concat(dolphs_dfs)
```

```python
bills_data = mega_Bills_df['clean_Tweet']
dolphs_data = mega_Dolphs_df['clean_Tweet']
```

```python
bills_tfIdfVectorizer=TfidfVectorizer(use_idf=True)
bills_tfIdf = bills_tfIdfVectorizer.fit_transform(bills_data)
bills_df_vect = pd.DataFrame(bills_tfIdf[0].T.todense(), index=bills_tfIdfVectorizer.get_feature_names_out(), columns=["TF-IDF"])
bills_df_vect = bills_df_vect.sort_values('TF-IDF', ascending=False)
bills_df_vect.head(25)
```

```python
dolphs_tfIdfVectorizer=TfidfVectorizer(use_idf=True)
dolphs_tfIdf = dolphs_tfIdfVectorizer.fit_transform(dolphs_data)
dolphs_df_vect = pd.DataFrame(dolphs_tfIdf[0].T.todense(), index=dolphs_tfIdfVectorizer.get_feature_names_out(), columns=["TF-IDF"])
dolphs_df_vect = dolphs_df_vect.sort_values('TF-IDF', ascending=False)
dolphs_df_vect.head(25)
```
