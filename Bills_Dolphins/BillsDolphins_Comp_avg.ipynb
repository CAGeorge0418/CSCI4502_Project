```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
bills_csvs = os.listdir('Bills_CSVs')
bills_dfs = [pd.read_csv('Bills_CSVs/'+csv) for csv in bills_csvs]
dolphins_csvs = os.listdir('Dolphins_CSVs')
dolphins_dfs = [pd.read_csv('Dolphins_CSVs/'+csv) for csv in dolphins_csvs]
```

```python
bills_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in bills_dfs]
bills_dfs_avg_comps_reordered = [bills_dfs_avg_comps[11],bills_dfs_avg_comps[12],bills_dfs_avg_comps[13],bills_dfs_avg_comps[2],bills_dfs_avg_comps[3],bills_dfs_avg_comps[4], bills_dfs_avg_comps[5], bills_dfs_avg_comps[6], bills_dfs_avg_comps[7], bills_dfs_avg_comps[8], bills_dfs_avg_comps[9], bills_dfs_avg_comps[10],bills_dfs_avg_comps[0], bills_dfs_avg_comps[1]]
bills_dfs_win_perc = [df.loc[df['Positive_Sentiment']==True].shape[0]/df.shape[0] for df in bills_dfs]
bills_dfs_win_percs_reordered = [bills_dfs_win_perc[11],bills_dfs_win_perc[12],bills_dfs_win_perc[13],bills_dfs_win_perc[2],bills_dfs_win_perc[3],bills_dfs_win_perc[4], bills_dfs_win_perc[5], bills_dfs_win_perc[6], bills_dfs_win_perc[7], bills_dfs_win_perc[8], bills_dfs_win_perc[9], bills_dfs_win_perc[10],bills_dfs_win_perc[0], bills_dfs_win_perc[1]]
bills_wl = ['win', 'win', 'loss', 'win', 'win', 'loss', 'win', 'win', 'win', 'loss', 'loss', 'win', 'win', 'win']

dolphins_dfs_avg_comps = [df['Compound_Sentiment_Score'].sum()/df.size for df in dolphins_dfs]
dolphins_dfs_avg_comps_reordered = [dolphins_dfs_avg_comps[11],dolphins_dfs_avg_comps[12],dolphins_dfs_avg_comps[13],dolphins_dfs_avg_comps[2],dolphins_dfs_avg_comps[3],dolphins_dfs_avg_comps[4], dolphins_dfs_avg_comps[5], dolphins_dfs_avg_comps[6], dolphins_dfs_avg_comps[7], dolphins_dfs_avg_comps[8], dolphins_dfs_avg_comps[9], dolphins_dfs_avg_comps[10],dolphins_dfs_avg_comps[0], dolphins_dfs_avg_comps[1]]
dolphins_dfs_win_perc = [df.loc[df['Positive_Sentiment']==True].shape[0]/df.shape[0] for df in dolphins_dfs]
dolphins_dfs_win_percs_reordered = [dolphins_dfs_win_perc[11],dolphins_dfs_win_perc[12],dolphins_dfs_win_perc[13],dolphins_dfs_win_perc[2],dolphins_dfs_win_perc[3],dolphins_dfs_win_perc[4], dolphins_dfs_win_perc[5], dolphins_dfs_win_perc[6], dolphins_dfs_win_perc[7], dolphins_dfs_win_perc[8], dolphins_dfs_win_perc[9], dolphins_dfs_win_perc[10],dolphins_dfs_win_perc[0], dolphins_dfs_win_perc[1]]
dolphins_wl = ['win', 'loss', 'win', 'win', 'win', 'win', 'loss', 'loss', 'loss', 'win', 'win', 'win', 'win', 'win']
```

```python
fig, (bills_plot_1, bills_plot_2) = plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(ax=bills_plot_1 , x=range(len(bills_dfs_avg_comps))+np.ones(len(bills_dfs_avg_comps)), y=bills_dfs_avg_comps_reordered, hue=bills_wl)
bills_plot_1.set_title('Bills Avg. Sentiment Per Game')
bills_plot_1.set_xlabel('Game Number')
bills_plot_1.set_xticks(range(len(bills_dfs_avg_comps))+np.ones(len(bills_dfs_avg_comps)))
bills_plot_1.set_ylabel('Comp. Sentiment')

sns.scatterplot(ax=bills_plot_2 , x=range(len(bills_dfs_win_perc))+np.ones(len(bills_dfs_win_perc)), y=bills_dfs_win_percs_reordered, hue=bills_wl)
bills_plot_2.set_title('Bills Positive Perc. Per Game')
bills_plot_2.set_xlabel('Game Number')
bills_plot_2.set_xticks(range(len(bills_dfs_win_perc))+np.ones(len(bills_dfs_win_perc)))
bills_plot_2.set_ylabel('Positive Perc.')
```

```python
fig, (dolphins_plot_1, dolphins_plot_2) = plt.subplots(1,2,figsize=(15,5))
sns.scatterplot(ax=dolphins_plot_1 , x=range(len(dolphins_dfs_avg_comps))+np.ones(len(dolphins_dfs_avg_comps)), y=dolphins_dfs_avg_comps_reordered, hue=dolphins_wl)
dolphins_plot_1.set_title('Dolphins Avg. Sentiment Per Game')
dolphins_plot_1.set_xlabel('Game Number')
dolphins_plot_1.set_xticks(range(len(dolphins_dfs_avg_comps))+np.ones(len(dolphins_dfs_avg_comps)))
dolphins_plot_1.set_ylabel('Comp. Sentiment')

sns.scatterplot(ax=dolphins_plot_2 , x=range(len(dolphins_dfs_win_perc))+np.ones(len(dolphins_dfs_win_perc)), y=dolphins_dfs_win_percs_reordered, hue=dolphins_wl)
dolphins_plot_2.set_title('Dolphins Positive Perc. Per Game')
dolphins_plot_2.set_xlabel('Game Number')
dolphins_plot_2.set_xticks(range(len(dolphins_dfs_win_perc))+np.ones(len(dolphins_dfs_win_perc)))
dolphins_plot_2.set_ylabel('Positive Perc.')
```
